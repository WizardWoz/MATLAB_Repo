
<samp>下面我们再来介绍</samp>MATLAB<samp>中两个使用频率很高的逻辑运算符：</samp>

## **&&****<samp>和</samp>****||**

<samp>这两个运算符和前面学的“逻辑与</samp>&<samp>”和“逻辑或</samp>|<samp>”非常相似，但它们有两个非常重要的区别：</samp>


<samp>（</samp>1<samp>）</samp>&&<samp>和</samp>||<samp>只能对标量（只有一个值）进行逻辑运算，不能对有多个元素的向量或者矩阵进行运算，而</samp>&<samp>和</samp>|<samp>可以。比如我们上面那个练习题，你只能使用</samp>&<samp>和</samp>|<samp>进行运算。</samp>


```matlab
% (A >= 60) && (A < 80)
```


<samp>（</samp>2<samp>）</samp>&&<samp>和</samp>||<samp>进行逻辑运算时具有短路功能，可以提高运行效率：</samp>

-  计算A && B时<samp>，如果</samp>A<samp>为逻辑</samp>0<samp>，则</samp>B<samp>不会被判断，因为最后的结果一定是逻辑</samp>0<samp>；</samp>
-  计算A || B时<samp>，如果</samp>A<samp>为逻辑</samp>1<samp>，则</samp>B<samp>不会被判断，因为最后的结果一定是逻辑</samp>1<samp>。</samp>

<samp>举个例子：假设</samp>a<samp>等于</samp>10<samp>，</samp>b<samp>等于</samp>3<samp>，现在要计算：</samp>(a+b < 10) && (a/b > 1)<samp>，那么</samp>MATLAB<samp>首先会判断前面一项：</samp>(a+b < 10)<samp>，因为这一项计算的结果为逻辑</samp>0<samp>，所以后面的</samp>(a/b > 1)<samp>这一项不会被计算，</samp>MATLAB<samp>会直接返回逻辑</samp>0<samp>；如果你使用的是：</samp>(a+b < 10) & (a/b > 1)<samp>，那么这两项都会被计算，这样的话效率会低一点。在下一章中，我们会介绍</samp>if<samp>判断语句和</samp>while<samp>循环语句，和</samp>&<samp>、</samp>|<samp>相比，</samp>&&<samp>和</samp>||<samp>在</samp>if<samp>和</samp>while<samp>语句中使用频率更高。</samp>


```matlab
a = 10;
b = 2;
(a+b < 10) & (a/b > 1)
```

```matlab
(a+b < 10) && (a/b > 1)
```


**<samp>思考题：</samp>**<samp>前面我们介绍过</samp>logical<samp>函数，它能将数值转换为逻辑值，但如果输入的是</samp>NaN<samp>，那么</samp>MATLAB<samp>会报错，请大家思考：分别运行</samp>(10 > 3) | logical(NaN)<samp>和</samp>(10 > 3) || logical(NaN)<samp>的结果是什么？</samp>


<samp>运行</samp> (10 > 3) | logical(NaN) <samp>会报错，因为前后两项都会被计算，所以当</samp>MATLAB<samp>运行到后面的</samp>logical(NaN)<samp>时会报错；运行</samp> (10 > 3) || logical(NaN) <samp>会返回逻辑</samp>1<samp>，因为前面一项（</samp>10>3<samp>）返回逻辑</samp>1<samp>，此时会触发</samp>||<samp>的短路机制，这时候就会直接返回逻辑</samp>1<samp>。</samp>


```matlab
% 报错：错误使用 logical NaN 值无法转换为逻辑值。
% (10 > 3) | logical(NaN)
(10 > 3) || logical(NaN)
```

## 利用逻辑值对矩阵进行索引

<samp>前面我们介绍过</samp>MATLAB<samp>中的向量和矩阵，并介绍过如何通过元素所处的位置来进行索引。</samp>


<samp>例如：</samp>

-  <samp>若</samp>a<samp>是一个向量，则</samp>a(1:2:end)<samp>表示提取</samp>a<samp>中所有奇数位置的元素；</samp>

```matlab
a = [10 1 4 4 1 10 2 1 7 8];
a(1:2:end)
```


-  <samp>若</samp>A<samp>是一个矩阵，则</samp>A(1:2:end, :)<samp>表示提取</samp>A<samp>中所有奇数行的元素。</samp>

```matlab
A = [3 4 4 10;
     7 10 9 3;
     4 2 6 3;
     6 3 4 8;
     9 9 9 2];
A(1:2:end, :)
```


**<samp>本小节我们介绍矩阵索引的另一种方式：利用逻辑值。</samp>**


<samp>假设有一个</samp><samp>m</samp><samp>行</samp><samp>n</samp><samp>列的矩阵</samp><samp>A</samp><samp>，我们要提取其指定位置的元素，那么我们可以生成一个和</samp><samp>A</samp><samp>同样大小的逻辑矩阵</samp><samp>L</samp><samp>，</samp><samp>L</samp><samp>中的元素要么是逻辑值</samp><samp>1</samp><samp>，要么是逻辑值</samp><samp>0</samp><samp>，其中：等于逻辑值</samp><samp>1</samp><samp>的元素所处的位置是我们所需要的。接着我们只需要使用命令</samp><samp>A(L)</samp><samp>，就能够在</samp><samp>A</samp><samp>中提取出指定位置的元素。通常逻辑矩阵</samp><samp>L</samp><samp>是由一系列的逻辑运算或者条件运算得来的。</samp>


```matlab
A = randi(10,3,4)
```

```matlab
L = (A <= 5)
```

```matlab
A(L)
```


这里的L实际上是我通过关系运算得到的一个逻辑矩阵，它的大小和A相同，L中等于1的位置就是矩阵A中要索引的位置。


注意：命令A(L)得到的结果是一个列向量，有同学会思考：为什么A(L)返回的结果的顺序是这样的，而不是其他的顺序。这是因为**在MATLAB中，矩阵的数据在计算机的内存中被存储为单列，由矩阵的各列顺次连接而成。因此这里返回的结果也是根据L中等于1的位置按照各列顺次连接的**。


```matlab
A(A<=5)
```

```matlab
A( A>=5 & A < 8)
```


**我们再来看一个有趣的问题：缺失值的识别和填补。**


<samp>举个例子：假设清风老师要连续一周测量早上</samp>6<samp>点室外的温度，结果清风老师周二和周五睡过了头，那两天的温度没有测量，剩下五天的温度分别是</samp>10<samp>°、</samp>5<samp>°、</samp>2<samp>°、</samp>8<samp>°和</samp>5<samp>°，周二和周五的温度成了缺失值。现在清风老师想利用有数据的剩余五天的平均气温来代替周二和周五这两天的温度，于是他计算出这五天的平均气温为（</samp>10+5+2+8+5<samp>）</samp>/5=6<samp>°，这时候就完成了对缺失值的填补。</samp>


<samp>那么我们怎样在</samp>MATLAB<samp>中实现这个过程呢？我们可以先定义一个向量</samp>A<samp>用来保存这一周的温度：</samp>A = \[10 NaN 5 2 NaN 8 5\]<samp>，其中第二个元素和第五个元素为</samp>NaN<samp>，代表周二和周五的温度数据是缺失的。现在需要大家将</samp>A<samp>中所有的</samp>NaN<samp>值替换成所有非缺失值的平均值。</samp>


```matlab
A = [10 NaN 5 2 NaN 8 5];
isnan(A)
```

```matlab
~(isnan(A))
```


答案只需要一行代码：


```matlab
A(isnan(A)) = mean(A(~isnan(A)))
```


**<samp>这里出现了</samp>****isnan****<samp>函数，它可以判断数组中的元素是否为不定值</samp>****NaN****<samp>，并返回一个和输入的数组大小相同的逻辑数组。</samp>**


<samp>例如，这里的</samp>isnan(A)<samp>返回的结果就是</samp>\[0 1 0 0 1 0 0\]<samp>这个逻辑数组。</samp>


```matlab
A = [10 NaN 5 2 NaN 8 5];
isnan(A)
```


<samp>有同学会想：为什么不直接用命令</samp>A==NaN<samp>来找</samp>A<samp>中的缺失值？</samp>


<samp>这是因为在</samp>MATLAB<samp>中，</samp>NaN<samp>相互之间不相等，运行</samp>NaN == NaN<samp>会返回逻辑</samp>0<samp>。</samp>


```matlab
% NaN == NaN
% A == NaN
```


下一个问题：如何找出A中所有不是缺失值的元素，我们可以对isnan(A)的结果进行逻辑非操作。


```matlab
~isnan(A)
```

```matlab
A(~isnan(A)) % 有数据的五天的温度
```

```matlab
mean(A(~isnan(A)))  % 计算平均值
```

## **all****<samp>、</samp>****any****<samp>和</samp>****find****<samp>函数</samp>**

<samp>下面我们介绍三个非常重要的函数，它们的功能如下表所示：</samp>





<samp>事实上，</samp>all<samp>函数和</samp>any<samp>函数的用法和我们之前讲解的</samp>sum<samp>函数非常像</samp>


```matlab
A = randi([-5,5],1,8)
```

```matlab
all(A)
```

```matlab
A = randi([-3,3],3,4)
```

```matlab
all(A)
```

```matlab
all(A,1)
```


<samp>思考：如果要判断整个矩阵的元素是否全部为非零值应该怎么办？</samp>


<samp>至少两种方法：</samp>all(A(:)) <samp>或者</samp>all(all(A))


```matlab
A = randi([0,2],2,2)
```

```matlab
all(A(:))
```

```matlab
all(all(A))
```


<samp>可以看出，</samp>all<samp>函数相当于对向量或者矩阵的元素进行‘逻辑与</samp>&<samp>’运算，只有全为非零值时才返回逻辑值</samp>1<samp>。而</samp>any<samp>函数则相当于对元素进行‘逻辑或</samp>|<samp>’运算，存在至少一个非零值时就会返回逻辑值</samp>1<samp>。</samp>


```matlab
A = [0 0 0 -1 0 20 0];
any(A)
```

```matlab
A = [0 0 0 0 0 0];
any(A)
```

```matlab
A = [0  0  -1  2;
    -3  0  1  -2;
    0  0  0  0];
any(A)
```

```matlab
any(A,1)
```

```matlab
any(A,2)
```


<samp>思考：如果要判断整个矩阵的元素是否至少含有一个非零值，应该怎么办？</samp>


<samp>至少两种方法：</samp>any(A(:)) <samp>或者</samp>any(any(A))


```matlab
A = randi([0,2],2,2)
```

```matlab
any(A(:))
```

```matlab
any(any(A))
```


<samp>事实上，</samp>all<samp>函数和</samp>any<samp>函数很少直接运用在数值矩阵上，它常常配合逻辑矩阵来实现特定的功能。</samp>


<samp>（</samp>2<samp>）三门科目的成绩都不低于</samp>85<samp>分的同学可以获得奖学金评选的资格，请指出哪些同学可以获得资格。要求返回一个包含</samp>100<samp>个元素的逻辑向量，元素为</samp>1<samp>的下标对应的同学有评选资格。</samp>


<samp>（4）这三门科目中是否有科目没有人挂科。要求返回一个包含3个元素的逻辑向量，元素为</samp>1<samp>的下标对应的科目表示没有人挂科。</samp>
